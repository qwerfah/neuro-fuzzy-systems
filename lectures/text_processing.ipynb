 ### Лекция (30.04.21)
 
`TF = term frequency` - частота термина (в зависимости от того, что используется)

`IDF = inverced document frequency`

В зависимости от задачи можно взять токены, словоформы, начальные формы. 

У нас есть коллекция документов. Для каждого можно посчитать частоту терминов TF (что бы не понималось под термом).

Слова типа "a" в английском не несут в себе никакого смысла, но указывают на то, что за ним будет к-л именная группа (терм).

Если слово (терм) встречается во всех документах в коллекции (или в большинстве), это важное слово для коллекции и может указывать на принадлежность текстов к определенной тематике.

Закон Ципфа: `fi = f0 / (betta^alphai)`

#### Почему TF и IDF лучше вектора частот:

Можно разбить документ на множество векторов

**Понятие коллокации:** лучшим определеинем является набор совместно стоящих слов.

MI = mutual information

`MI = (p(a)p(b) / p(ab))^-1`

Чем чаще два слова встречаются вместе, тем больше знаменатель. Чем реже они встречаются раздельно, тем меньше числитель. Таким образом, чем чащи они вместе и чем реже они раздельно, тем больше информации.

MI считается для словосочетаний для которых частота встречаемости не меньше некоторой заданной.

Помимо MI есть log-likelyhood (поощряет длинные сочетания), t-score (поощряет авторскую лексику, которая встречается часто, но в одном тексте).

Инструменты для работы с разреженными матрицами встроены в skilearn.

Для текста на частотах косинусоидальная метрика не рабоатет, на нормированных данных - да.
